### Store Part I results into DataFrame

```python
file = "output_data/cities.csv"

cities = pd.read_csv(file)

cities.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
# Source: https://jupyter-gmaps.readthedocs.io/en/latest/tutorial.html
gmaps.configure(api_key = g_key)

# define locations and weight
locations = cities[["Latitude", "Longitude"]]
weight = cities["Humidity"]
```

```python
# generate figure
fig = gmaps.figure(center = (15, 0), zoom_level = 2)

# add heatmap layer
heatmap_layer = gmaps.heatmap_layer(locations, weight)
fig.add_layer(heatmap_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
ideal_cities = cities.loc[(cities["Max Temperature"] > 60) & (cities["Max Temperature"] < 76) &
                         (cities["Cloudiness"] < 50) & (cities["Humidity"] < 50) & (cities["Wind Speed"] < 8)].dropna()

ideal_cities
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotels = ideal_cities

# add hotel name column
hotels["Hotel Name"] = ""

hotels
```

```python
# set up log file print statement
print(f"Beginning Data Retrieval")
print(f"------------------------")


# loop through rows in list
for index, row in hotels.iterrows():
    
    # set up parameters
    lat = row["Latitude"]
    lng = row["Longitude"]
    city = row["City"]
    location = f"{lat}, {lng}"
    
    params = {
        "location": location,
        "radius": 5000,
        "types": "hotel",
        "keyword": "hotel",
        "key": g_key
    }
    
    # set up call
    call_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
    
    response = requests.get(call_url, params=params).json()
    output = response['results']
        
    # try statement to append calls where a value is found
    try:
        print(f"Retrieving hotel for {city}")
        print(f"Hotel Name: {output[0]['name']}")
        hotels.loc[index, "Hotel Name"] = output[0]['name']
        print("------------")
    except:
        print(f"No results. Skipping...")
        

print("--------Complete--------")
```
